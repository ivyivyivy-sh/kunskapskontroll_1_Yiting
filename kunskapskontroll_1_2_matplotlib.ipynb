### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

In Matplotlib, there are two main styles for creating plots: Object-Oriented (OO) Style and Pyplot (State-Based) Style. 

Matplotlib's documentation and examples use both the OO and the pyplot styles. In general, we suggest using the OO style, particularly for complicated plots, and functions and scripts that are intended to be reused as part of a larger project. However, the pyplot style can be very convenient for quick interactive work.


### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure: The overall container for everything in a plot, acting as a "canvas" for the visual elements. A figure can contain multiple plots (axes).

Axes: The actual area where data is plotted within a figure. It includes elements like titles, x-axis, and y-axis labels, and allows for specific data plotting and customization.

Axis: Refers to the x-axis or y-axis within an Axes. It handles the scale, limits, and tick marks along the respective axis.

Artist: Any visual element in Matplotlib, including the figure, axes, lines, text, and labels. Artists are the building blocks that make up the visual representation of the data.

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 100) 
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x**2, label='quadratic')
ax.set_xlabel('x label')
ax.set_ylabel('y label') 
ax.set_title("Simple Plot") 
ax.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-4, 4, 100) 
plt.figure(figsize=(5, 2.7), layout='constrained')
plt.plot(x, x**2, label='quadratic')  
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Scatter plot in the first subplot
ax1.scatter(random_data_x, random_data_y, alpha=0.5, color='blue', label="Random Data")
ax1.plot(x, y, color='red', label="y = x^2")
ax1.set_title("Scatter Plot")
ax1.set_xlabel("X-axis")
ax1.set_ylabel("Y-axis")
ax1.legend()

# Bar plot in the second subplot
ax2.bar(names, values, color='green')
ax2.set_title("Bar Plot")
ax2.set_xlabel("Fruit")
ax2.set_ylabel("Quantity")

# Display the plots
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
